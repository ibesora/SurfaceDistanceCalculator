```python
import numpy as np
import matplotlib.pyplot as plt
from calculator import computeSurfaceDistance
```

```python
rawData = np.reshape(np.fromfile('data/pre.data', dtype=np.ubyte), (512, 512))
rawData2 = np.reshape(np.fromfile('data/post.data', dtype=np.ubyte), (512, 512))
```

```python
startPoint = (250, 0)
endPoint = (250, 511)
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(20, 20))
ax1.imshow(rawData, cmap='gray', vmin=0, vmax=255)
ax1.plot([startPoint[0], endPoint[0]], [startPoint[1], endPoint[1]])
ax2.imshow(rawData2, cmap='gray', vmin=0, vmax=255)
ax2.plot([startPoint[0], endPoint[0]], [startPoint[1], endPoint[1]])
```

```python
results = computeSurfaceDistance(startPoint, endPoint, rawData, 30, 11)
results2 = computeSurfaceDistance(startPoint, endPoint, rawData2, 30, 11)
heights = [results[i]["height"] for i in range(0, len(results))]
distances = [results[i]["accumDistance"] for i in range(0, len(results))]
surfaceDist = results[len(results) - 1]["accumSurfaceDistance"]
heights2 = [results2[i]["height"] for i in range(0, len(results2))]
distances2 = [results2[i]["accumDistance"] for i in range(0, len(results2))]
surfaceDist2 = results2[len(results2) - 1]["accumSurfaceDistance"]
plt.figure(figsize=(25,5))
plt.plot(distances, heights)
plt.plot(distances2, heights2)
```

```python
print("Distance (m) pre eruption:", surfaceDist)
print("Distance (m) post eruption:", surfaceDist2)
print("Difference (m):", surfaceDist - surfaceDist2)
print("Number of steps:", len(results))
```
